```python
import pandas as pd
import matplotlib.pyplot as plt

# Muat file CSV
df = pd.read_csv('products.csv')

# Konfigurasi plot
plt.figure(figsize=(12, 5))

# --- 1. Histogram untuk Harga (Price) ---
plt.subplot(1, 2, 1) # Membuat subplot di posisi 1
# Menggunakan 15 bins untuk memecah data menjadi 15 kelompok
df['price'].hist(bins=15, edgecolor='black', color='skyblue')
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga (Price)')
plt.ylabel('Frekuensi')
plt.grid(axis='y', alpha=0.75)


# --- 2. Histogram untuk Kuantitas (Quantity) ---
plt.subplot(1, 2, 2) # Membuat subplot di posisi 2
df['quantity'].hist(bins=20, edgecolor='black', color='lightcoral')
plt.title('Distribusi Kuantitas Stok')
plt.xlabel('Kuantitas (Quantity)')
plt.ylabel('Frekuensi')
plt.grid(axis='y', alpha=0.75)

# Tampilkan plot
plt.tight_layout()
plt.show()
```
